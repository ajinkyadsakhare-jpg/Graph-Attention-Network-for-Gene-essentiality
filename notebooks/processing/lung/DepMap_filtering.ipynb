```python
import pandas as pd 
import numpy as np
```

```python
model = pd.read_csv("../../../data/raw/depmap/Model.csv")
model
```

```python
model[model["OncotreeLineage"]=='Lung']
```

```python
lung_id = model[model["OncotreeLineage"] == 'Lung']['ModelID']
lung_id
```

```python
expressions_data = pd.read_csv("../../../data/raw/depmap/OmicsExpressionTPMLogp1HumanProteinCodingGenes.csv")
expressions_data
```

```python
expressions_data = expressions_data[expressions_data['IsDefaultEntryForModel'] == 'Yes']
```

```python
expressions_data.shape
```

```python
lung_expression = expressions_data[expressions_data['ModelID'].isin(lung_id)]
lung_expression
```

```python
lung_expression.drop(['Unnamed: 0', 'SequencingID', 'IsDefaultEntryForModel', 'ModelConditionID', 'IsDefaultEntryForMC'], axis=1, inplace=True)
```

```python
lung_expression
```

```python
lung_expression.columns = ['ModelID'] + list(lung_expression.columns[1:].str.split(" (", regex=False).str[0])
```

```python
ppi = pd.read_csv('../../../data/lung/lung_PPI_final.csv')
ppi_genes = pd.concat([ppi['A'], ppi['B']]).unique()
```

```python
gene_cols = lung_expression.columns[1:][lung_expression.columns[1:].isin(ppi_genes)]
```

```python
lung_expression = lung_expression[['ModelID'] + list(gene_cols)]
```

```python
lung_expression.shape
```

```python
lung_expression.to_csv('../../../data/lung/Lung_expression_final.csv', index=False)
```
